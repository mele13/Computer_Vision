TAREA: Crea una imagen, p.e. 800x800, con la textura del tablero de ajedrez

```python
tamano_tablero_pixeles = 800
tamano_cuadrado_pixeles = tamano_tablero_pixeles // 8
grosor_borde = 2 

#Creación tablero
tablero = np.zeros((tamano_tablero_pixeles,tamano_tablero_pixeles, 3), np.uint8) 

# RGeneración de cuadrados blancos
for i in range(8):
    for j in range(8):
        if (i + j) % 2 == 0:
            tablero[i*tamano_cuadrado_pixeles+grosor_borde:(i+1)*tamano_cuadrado_pixeles-grosor_borde, j*tamano_cuadrado_pixeles+grosor_borde:(j+1)*tamano_cuadrado_pixeles-grosor_borde] = [255, 255, 255]

#Mostrar la imagen
plt.imshow(tablero)

# Modificar etiquetas
letras = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H']
numeros = ['8', '7', '6', '5', '4', '3', '2', '1']
plt.xticks(np.arange(tamano_cuadrado_pixeles/2, tamano_tablero_pixeles, tamano_cuadrado_pixeles), letras)
plt.yticks(np.arange(tamano_cuadrado_pixeles/2, tamano_tablero_pixeles, tamano_cuadrado_pixeles), numeros)

# Eliminar Ticks
plt.tick_params(axis='both', bottom=False, left=False)
plt.title('Tablero de Ajedrez')
plt.axis('on')
#Mostrar el tablero
plt.show()
```

TAREA: Crear una imagen estilo Mondrian (un ejemplo https://www3.gobiernodecanarias.org/medusa/ecoescuela/sa/2017/04/17/descubriendo-a-mondrian/ )

```python
# Tamaño del lienzo
altura_lienzo, ancho_liezo = 800, 800
borde = 5

# Crear un lienzo
lienzo = np.zeros((altura_lienzo, ancho_liezo, 3), dtype=np.uint8)

# Número aleatorio de divisiones horizontales y verticales
num_divisiones_horizontales = random.randint(3, 10)
num_divisiones_verticales = random.randint(3, 10)

# Colores primarios de Mondrian
colores_modrian = [(0, 0, 0), (255, 255, 255), (255, 0, 0), (0, 255, 255), (0, 0, 255)]

# Lista de anchos de las columnas
anchos_columnas = [random.randint(50, 150) for _ in range(num_divisiones_horizontales)]

# Lista de alturas de cada cuadrado
alturas_filas = [[random.randint(50, 150) for _ in range(num_divisiones_horizontales)] for _ in range(num_divisiones_verticales)]

# Dibujar cuadrados
for i, altura_fila in enumerate(alturas_filas):
    for j, ancho_columna in enumerate(anchos_columnas):
        y1 = sum([alturas_filas[k][j] for k in range(i)]) + borde        
        y2 = y1 + altura_fila[j] if i != num_divisiones_verticales -1 else altura_lienzo 

        x1 = sum(anchos_columnas[:j]) + borde        
        x2 = x1 + ancho_columna if j != num_divisiones_horizontales - 1 else ancho_liezo 

        random.shuffle(colores_modrian)
        if colores_modrian:
            # Toma el primer color de la lista si hay colores disponibles
            color = colores_modrian.pop()  
        else:
            # Inicializamos de nuevo
            colores_modrian = [(0, 0, 0), (255, 255, 255), (255, 0, 0), (0, 255, 255), (0, 0, 255)]
            color = colores_modrian.pop()  
        
        cv2.rectangle(lienzo, (x1, y1), (x2, y2), color, -1)
        cv2.rectangle(lienzo, (x1, y1), (x2, y2), (0, 0, 0), borde)

# Muestra el lienzo
cv2.imshow('Cuadro de Mondrian', lienzo)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

TAREA: Modifica de alguna forma los valores de un plano de la imagen

```python
#Carga imagen desde disco
img = cv2.imread('./Images/otter-21004.jpg') 
#Dimensiones
w, h = img.shape[:2]
  
#Tamaño de la nueva imagen
xNew = int(w * 1 / 4)
yNew = int(h * 1 / 4)

#Crear matriz para la imagen recortada
newImage = np.zeros([xNew, yNew, 3])
  
#Copiar pixeles desde la posición indicada (400,400), de la imagen original
for i in range(xNew):
    for j in range(yNew):
        newImage[i, j]= img[400 + i, 400 + j]

#Convertir la nueva imagen a valores válidos
newImage = cv2.convertScaleAbs(newImage)
#Mostrar imagen recortada
plt.imshow(newImage) 
plt.axis('off')
plt.show()

#Rotar la imagen 180º
matriz_rot = cv2.getRotationMatrix2D((h/2, w/2), 180, 1)
imagen_rotada = cv2.warpAffine(img, matriz_rot, (h, w))

#Mostrar la imagen rotada
plt.imshow(imagen_rotada) 
plt.axis('off')
plt.show()

#Invertir los colores de la imagen
imagen_invertida = cv2.bitwise_not(img)

#Mostrar la imagen invertida
plt.imshow(imagen_invertida) 
plt.axis('off')
plt.show()
```

```python
# Cargar la imagen a color
img = cv2.imread('./Images/fox-715588_1280.jpg')

# Fuerza la lectura en escala de grises
gray_img = cv2.imread('./Images/fox-715588_1280.jpg', cv2.IMREAD_GRAYSCALE)

# Encuentra el píxel más oscuro y el más claro
coordenadas_oscuro = np.unravel_index(np.argmin(gray_img), gray_img.shape)
coordenadas_claro = np.unravel_index(np.argmax(gray_img), gray_img.shape)

# Destacar el píxel más oscuro
img_oscuro = img.copy()
cv2.circle(img_oscuro, (coordenadas_oscuro[1], coordenadas_oscuro[0]), 10, (0, 0, 0), -1)
cv2.circle(img_oscuro, (coordenadas_oscuro[1], coordenadas_oscuro[0]), 10, (255, 255, 255), 5)

# Destacar el píxel más claro
img_claro = img.copy()
cv2.circle(img_claro, (coordenadas_claro[1], coordenadas_claro[0]), 10, (255, 255, 255), -1)
cv2.circle(img_claro, (coordenadas_claro[1], coordenadas_claro[0]), 10, (0, 0, 0), 5)

# Mostrar imágenes
plt.imshow(cv2.cvtColor(img_oscuro, cv2.COLOR_BGR2RGB))
plt.title('Píxel más Oscuro')
plt.axis('off')
plt.show()

plt.imshow(cv2.cvtColor(img_claro, cv2.COLOR_BGR2RGB))
plt.title('Píxel más Claro')
plt.axis('off')
plt.show()
```

```python
# Función que trata eventos de ratón
def mouse_events(event, x, y, flags, params):
    global px, py, R, G, B
    px = -1
    # Movimiento del puntero
    if event == cv2.EVENT_MOUSEMOVE:
        # Valores del píxel
        B = frame[y, x, 0]
        G = frame[y, x, 1]
        R = frame[y, x, 2]

        px = x
        py = y

# Inicializar la captura de video
vid = cv2.VideoCapture(0)

lanzado = 0
px = -1

while True:
    #fotograma a fotograma
    ret, frame = vid.read()

    if ret:
        # Activa el manejador en el primer fotograma
        if lanzado == 0:
            # Muestra fotograma
            cv2.imshow('Cam', frame)
            # Define el nombre del manejador del evento
            cv2.setMouseCallback('Cam', mouse_events)
            lanzado = 1

        # Crea un degradado de color
        if px > -1:
            h, w = frame.shape[:2]
            degradado = np.zeros((h, w, 3), dtype=np.uint8)
            degradado[:, :, 0] = B
            degradado[:, :, 1] = G
            degradado[:, :, 2] = R
            # Nivel de transparencia
            alpha = 0.7  

            distancia_entre_circulos = 10

            for i in range(0, h, distancia_entre_circulos):
                for j in range(0, w, distancia_entre_circulos):
                    if (i // distancia_entre_circulos + j // distancia_entre_circulos) % 2 == 0:
                        cv2.circle(degradado, (j, i), 2, (0, 0, 0), -1)

            # Mezclar la imagen de la cámara con el degradado
            imagen_con_degradado = cv2.addWeighted(frame, alpha, degradado, alpha, 0)

            # Mostrar la imagen resultante
            cv2.imshow('Cam', imagen_con_degradado)
        else:
            cv2.imshow('Cam', frame)

    # Detenemos pulsado ESC
    if cv2.waitKey(20) == 27:
        break

# Libera el objeto de captura
vid.release()
# Destruye ventanas
cv2.destroyAllWindows()
```

Resaltado de oscuros y claros utilizando la cámara

```python
# Iniciar cámara
cap = cv2.VideoCapture(0)

while True:
    # fotograma a fotograma
    ret, frame = cap.read()

    if ret:
        # Encuentra el píxel más oscuro y el más claro
        coordenadas_oscuro = np.unravel_index(np.argmin(frame), frame.shape)
        coordenadas_claro = np.unravel_index(np.argmax(frame), frame.shape)

        # Destacar el píxel más oscuro
        img_oscuro = frame.copy()
        cv2.circle(img_oscuro, (coordenadas_oscuro[1], coordenadas_oscuro[0]), 10, (0, 0, 0), -1)
        cv2.circle(img_oscuro, (coordenadas_oscuro[1], coordenadas_oscuro[0]), 10, (255, 255, 255), 5)

        # Destacar el píxel más claro
        img_claro = frame.copy()
        cv2.circle(img_claro, (coordenadas_claro[1], coordenadas_claro[0]), 10, (255, 255, 255), -1)
        cv2.circle(img_claro, (coordenadas_claro[1], coordenadas_claro[0]), 10, (0, 0, 0), 5)

        # Combinar las imágenes en una sola
        combined_image = cv2.addWeighted(img_oscuro, 0.5, img_claro, 0.5, 0)

        # Mostrar la imagen con círculos
        cv2.imshow('Círculos Oscuro y Claro', combined_image)

    # Detenemos pulsado ESC
    if cv2.waitKey(20) == 27:
        break

# Libera el objeto de captura
cap.release()
# Destruye ventanas
cv2.destroyAllWindows()
```
